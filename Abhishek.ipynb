```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('virat.jpg', 1)
img.shape
```

```python
plt.imshow(img)
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
img_gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray)
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.imshow(img_invert)
blur_img = cv2.GaussianBlur(img_invert, (21, 21), 0) #
plt.title("Blurred Image")
```

```python
invert_blur_img = cv2.bitwise_not(blur_img)
plt.imshow(invert_blur_img)
plt.title("Inverted Blurred Image")
```

```python
sketch_img = cv2.divide(img_gray, invert_blur_img, scale=250.0)
plt.imshow(sketch_img)
plt.axis('off')
plt.title("Pencil Sketched Image").set_color("purple")
```
